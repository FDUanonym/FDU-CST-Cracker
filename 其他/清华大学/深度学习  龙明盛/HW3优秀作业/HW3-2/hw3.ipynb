# Plot curve

```python
import re

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_theme(font_scale=1.25)
```

```python
def plot(model_name, prog, columns, bbox_to_anchor=(0.5, 0.0, 0.5, 0.5), **kwargs):
    filename = f"{model_name.lower()}.log"
    with open(filename, encoding="utf-8") as f:
        data = [
            dict(zip(columns, prog.match(line).groups()))
            for line in f.readlines()
            if line
        ]
    df = pd.DataFrame(data, dtype=float).convert_dtypes()
    df.plot(x=columns[0], y=columns[1:], title=model_name, legend=False, **kwargs)
    plt.figlegend(loc="center", bbox_to_anchor=bbox_to_anchor)
    plt.savefig(f"{model_name}.png", dpi=300, bbox_inches="tight")
```

## DC-GAN implementation

```python
import torch
import torch.nn as nn
```

```python
class Flatten(nn.Module):
    def forward(self, x):
        N, C, H, W = x.size()  # read in N, C, H, W
        return x.view(
            N, -1
        )  # "flatten" the C * H * W values into a single vector per image


class Unflatten(nn.Module):
    """
    An Unflatten module receives an input of shape (N, C*H*W) and reshapes it
    to produce an output of shape (N, C, H, W).
    """

    def __init__(self, N=-1, C=128, H=7, W=7):
        super(Unflatten, self).__init__()
        self.N = N
        self.C = C
        self.H = H
        self.W = W

    def forward(self, x):
        return x.view(self.N, self.C, self.H, self.W)
```

```python
pooling = nn.MaxPool2d(kernel_size=2, stride=2)
conv1 = nn.Conv2d(
    in_channels=1,
    out_channels=32,
    kernel_size=4,
    stride=2,
    padding=1,
    bias=False,
)
conv2 = nn.Conv2d(
    in_channels=32,
    out_channels=64,
    kernel_size=3,
    stride=2,
    padding=1,
    bias=False,
)
fc1 = nn.Linear(64 * 2 * 2, 128)
fc2 = nn.Linear(128, 1)
```

```python
x = torch.rand(1, 1, 28, 28)
x = conv1(x)
print(x.size())
x = pooling(x)
print(x.size())
x = conv2(x)
print(x.size())
x = pooling(x)
print(x.size())
x = Flatten()(x)
print(x.size())
x = fc1(x)
print(x.size())
x = fc2(x)
print(x.size())
```
